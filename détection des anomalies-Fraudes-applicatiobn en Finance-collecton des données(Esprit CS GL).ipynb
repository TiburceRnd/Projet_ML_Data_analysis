```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X1 = np.random.randn(100000) + 10 #loi normal centré reduite
X2 = np.random.randn(100) + 200

X = list(X1) + list(X2)
X = np.array(X) 
plt.hist(X,bins = 1000);#ecartype a peut pres a egal 1
plt.grid()
```

```python
# On va générer des données selon une loi de probabilité 
# Générateur des nombres aleatoires

X1 = np.random.randn(100000) + 10 #loi normal centré reduite
X2 = np.random.randn(100) + 200

X = list(X1) + list(X2)
X = np.array(X) 
plt.hist(np.log10(X),bins = 1000,log = True);#ecartype a peut pres a egal 1
plt.grid()
```
